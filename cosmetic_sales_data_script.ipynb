```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
conn = sqlite3.connect("cosmetic_sales_data.db")
query = """
SELECT 
    product, 
    SUM(quantity) AS total_qty, 
    SUM(quantity * price) AS revenue 
FROM 
    sales 
GROUP BY 
    product
"""
df = pd.read_sql_query(query, conn)
conn.close()
print("Cosmetic Sales Summary:")
print(df)
df.plot(kind='bar', x='product', y='revenue', color='orchid', title='Revenue by Cosmetic Product', legend=False)
plt.ylabel('Revenue ($)')
plt.tight_layout()
plt.savefig("cosmetic_sales_chart.png")
plt.show()
```
